```python
df.isnull().sum() #checks if there is null value or not and if there it returns sum of it
df.dropna(axis=0,inplace=True)#removes the null value row from the dataset 
df.isnull().sum()
df.info()
```

Feature Selection Part

```python
df.columns
```

```python
df=df[['id','title','genre','original_language','overview',]]#only the columns listed here is preserve other are remove from df
```

```python
df
```

```python
df.original_language.value_counts()#counts the total number of data on the basis of ol
```

```python
df['tags']=df['overview']+df['genre']#this creates new column merging overview and genre
```

```python
df['language_tags']=df['original_language']+df['genre']#another column merging ol and genre
```

```python
df.head()
```

```python
new_dataset=df.drop(columns=['overview','genre','original_language'])#drops listed column and store new dataset, original df is unchanged
```

```python
new_dataset
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=9985, stop_words='english')
```

```python
cv1=CountVectorizer(max_features=9985, stop_words='english')
```

```python
cv
```

```python
cv1
```

```python
vtag=cv.fit_transform(new_dataset['tags'].values.astype('U')).toarray()#convert to unicode strings and convert to array
```

```python
vtag.shape
```

```python
vltag=cv1.fit_transform(new_dataset['language_tags'].values.astype('U')).toarray()
```

```python
vltag.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vtag)
```

```python
similarity1=cosine_similarity(vltag)
```

```python
similarity
```

```python
similarity1
```

```python
new_dataset[new_dataset['title']=="The Godfather"].index[0]
```

```python
distance =sorted(list(enumerate(similarity[2])),reverse =True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_dataset.iloc[i[0]].title)
```

```python
# it recmoomeds on the basis of Genre
def recommand(movies):
    index=new_dataset[new_dataset['title']==movies].index[0]
    distance =sorted(list(enumerate(similarity[index])),reverse =True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_dataset.iloc[i[0]].title)
```

```python
recommand('Dersu Uzala')#it is russian action/adventure movie and it recommends same genre movies
```

```python
#it recommends on the basis of Original Language
def recommand1(movies):
    # find the index of the movie in the dataset
    index=new_dataset[new_dataset['title']==movies].index[0]
    #calculate the similarity distance
    distance1 =sorted(list(enumerate(similarity1[index])),reverse =True, key=lambda vector:vector[1])
    #print the top 5 recommended movie titles
    for i in distance1[0:5]:
        print(new_dataset.iloc[i[0]].title)#iloc  get the row at the specified index and prints the movie title.
```

```python
recommand1('My Name Is Khan')#it is drama/romance hindi movie and it recommends the hindi movies from the same genre
```
